## YARIŞAN GRAFİKLER 

```python
#yeniden yükle
df=pd.read_csv("corona_dat.csv")
```

```python
df["date"].dtype
```

```python
import datetime
df["date"]=pd.to_datetime(df["date"])
df["date"].dtype
```

```python
df.info()
```

```python
df["month"]=df["date"].dt.month
df["year"]=df["date"].dt.year
df
```

```python
#df.drop(["month", "year"], axis=1, inplace=True)
#df
```

```python
df_yarisan_grf=df[["date", "Turkey", "Italy", "Spain", "Germany", "China", "Russia", "US"]]
```

```python
df_yarisan_grf.head()
```

```python
#index tarih sırasına göre girilmiş ama her zaman böyle olmayabilir, o yüzden datetime formatına çevirilir
#ve tarihe göre sıralama (index) yapılır 
df_yarisan_grf.set_index("date", inplace=True)
df_yarisan_grf.head()
```

```python
import seaborn as sns
#günlük vaka sayısı 
sns.lineplot(data=df_yarisan_grf);
```

```python
df_yarisan_grf=df_yarisan_grf.cumsum(axis=0)
df_yarisan_grf
```

```python
sns.lineplot(data=df_yarisan_grf, marker="o")
plt.xticks(rotation=90)
plt.title("Günlük vaka sayısı");
plt.show()
```
